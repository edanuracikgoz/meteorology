```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("data1.csv",delimiter=';')
df.columns =["year", "month", "day","hour","direction","velocity"]
df.velocity=df["velocity"]*3.6
print(df.head(5))
print(df.dtypes)
```

```python
##Sinop Meteoroloji stasyonu 2015-2020 Hz Verileri
## Grafiin altna saat skalasn ekle!!!
import datetime
import plotly.express as px
Ay = int(input("Please enter month number:\n"))
start=int(input("Please enter start day:\n"))
end=int(input("Please enter end day:\n"))

#yil=2014
yil=min(df.year)
maxyil=2020

first = df.loc[(df.year>=yil) & (df.year<=maxyil) &(df.month==Ay)& (df.day>=start)&(df.day<=end)]
    #const=df.loc[(df.year==min(df.year)) & (df.month==Ay)]
second=pd.DataFrame(first)
second.drop(['direction', 'velocity'], axis=1, inplace=True)
    #second.loc[:,'year'] = 2017
second= second.assign(year=2021)
secondtime=pd.to_datetime(second,format='%d-%m-%Y', errors='coerce')
first["datetime"]=secondtime

everynvelocity=first.velocity
dayrepeat=first.datetime
    #everynvelocity=first.groupby(everynvelocity.index // N).sum()/N
    #dayrepeat=dayrepeat[dayrepeat.index % N==0]
    
if (len(dayrepeat)==len(everynvelocity)):
    fig = px.scatter(first, x=dayrepeat, y=everynvelocity, color="year")
else :
    everynvelocity =everynvelocity[:-1]
    fig = px.scatter(first, x=dayrepeat, y=everynvelocity, color="year")   

fig.show()

```
